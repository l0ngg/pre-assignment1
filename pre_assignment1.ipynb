```
## problem 1 - folding a piece of paper
## calculate using exponent

#manual n for testing
#n = input()

n = 43
THICKNESS = 0.00008 # meters

folded_thickness = THICKNESS* 2**int(n)
print( "Thickness: {} meters".format(folded_thickness) )
```

```
# problem 2 - Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
```

```
## problem 3 - calculate using for loop

folded_thickness = 0.00008
for i in range(43):
  folded_thickness = folded_thickness * 2
print(folded_thickness , "meters")
```

```
## problem 4 - comparision for calculating time between the two method

import time
THICKNESS = 0.00008
folded_thickness = THICKNESS
n = input("enter the number of time you want the piece of paper folded : ")

##using for loop
start = time.time()
for i in range(int(n)):
  folded_thickness = folded_thickness * 2
elapsed_time = time.time() - start
print("\nloop function took : {}[s]".format(elapsed_time))

##using exponent function
start = time.time()
folded_thickness = THICKNESS* 2**int(n)
elapsed_time = time.time() - start
print("\nexponent function took : {}[s]".format(elapsed_time))

#exponent is faster
```

```
## problem 5 -- saving the values to a list

lst = []
folded_thickness = 0.00008
lst.append(folded_thickness)

for i in range(43):
  folded_thickness = folded_thickness * 2
  lst.append(folded_thickness)
#print(folded_thickness , "meters\n")

print("the list :", lst)
print(len(lst))
```

```
## problem 6 -- line graph

import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(lst)
plt.show()

# discussion:
print("even though we started with a thin piece of paper,\nwhen folded the thickness of the paper increase drastically, getting significantly larger and larger \nWhen showed on the graph we can see a exponential growth of the thickness of the paper")
```

```
## problem 7 -- cuztomizing graphs

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")

#red line
plt.plot(lst, color='red')
plt.show()

#o dots
plt.plot(lst, 'o')
plt.show()

#dot points connected by a line
plt.plot(lst, marker='o', color='cyan', linestyle='-')
plt.show()

#changing the axis value
plt.tick_params(labelsize=20) # Make settings related to axis values
plt.plot(lst)
plt.show()
```
